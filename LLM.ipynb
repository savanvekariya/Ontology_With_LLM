```python
test_sentence = """
In this project, we have addressed the topic of how improving the automotive electrical system can combat this problem by
measurably reducing fuel costs. First, a fundamental overview of the automotive electrical system is given. Each of the major
power-related elements of the automotive electrical system are reviewed. The project involves in addressing the problems above by
adding extra battery bank by a conjunction of PV array with MPPT [3] and hence decreasing the load on the alternator and
supplying sleep mode currents and loads with power generated for solar and ultimately reducing fuel consumption. A battery
management scheme for battery protection is developed.

Automotive Electrical System (AES) is a combination of various devices belonging to different domains. So, we need to analyze
each component individually to get an overall understanding about its parameters and their prominence. For ease of modelling, the
whole AES is divided into atomic units like the Alternator model, Starter motor, Lead-acid battery, PV array, MPPT, and DC-DC
converters.
The two components that generate power are the rotor and the stator. The rotor spins past three stationary stator windings, or wire
coils, around a fixed iron core that makes up the stator as the engine rotates the alternator pulley. A three-phase current is what it's
called. The coil windings are equally placed around the iron shaft at 120-degree intervals. The rotor's alternating magnetic field
causes an alternating current to flow in the stator. The AC current is delivered into a connected set of diodes via stator leads. To
manage the current, two diodes are connected to each stator lead. The diodes are used to stop and direct the flow of current. Because
batteries require DC electricity, the diodes act as a one-way valve, allowing current to flow only in one direction.

A starter, also known as a starter motor it is generally a DC motor, it is an electrical device that rotates (cranks) internal combustion
engines to allow them to start under their own power. Initially starter motor gets supply from 12V Lead Acid battery it draws huge
amount of current. Later the engine was detached from the starter motor as soon as it started running and now relies on the
combustion process. The starter motor gear meets the teeth of the flywheel in this component, which is located on the engine's
gearbox housing. Based on engine power starter motor power rating varies, in this we have used 5KW motor. 
The battery is modelled as a series resistor and a constant voltage source if the Battery charge capacity parameter is set to infinite.
The block mimics the battery as a series resistor and a charge-dependent voltage source if the Battery charge capacity parameter is
set as Finite. In this we have considered battery with no dynamics, no self-discharge, and with no aging.
The mathematical equations of voltage and current are used and modelled using MATLAB/Simulink for PV array modelling. Solar
cell modelling equations are simply multiplied by factors of series and parallel no of cells because an array is a series-parallel
combination of solar cells. The single-diode electrical equivalent circuit of a PV cell is shown in Figure 5 is sufficient for PV Array
modelling. It includes a current source (Iph) in parallel with a diode (D), a shunt resistance (Rsh) and a resistance (Rs) in series.
Buck converter is a power electronic circuit that steps down the input DC voltage to the output load. Thus, the average output
voltage of the buck converter is smaller than the source voltage. Buck converter circuit is shown in fig6. The working principle of
the circuit depends on the condition of the switch. The switch has two states, which are closed (ON) or open (OFF). Switch (S) is
generally a power electronic component such as MOSFETs. The operating mode buck converter consists of two types of continuous
and discontinuous. Circuit operating modes depend on the current through the inductor. In continuous mode, the inductor current
never reaches zero while in discontinuous mode will reach a value of zero for a time.
Engine averaged model is developed with the help of parameters like car tire diameter and we have given the speed manually. The
engine subsystem is shown in below fig. 8 it is the model for starting, idle and also for some constant speed which is same for both
cases with PV and without PV.
To start the engine in IC Engines it requires starting system which contains clutch, Starter solenoid, Pinion, starter motor etc. By
pressing clutch the plunger will move the pinion gear flywheel teeth which is located on starter motor shaft will attached to the
engine crank shaft later when we turn the ON the key solenoid coils get supply from the battery and will acts as a switch to the
starter motor. So, when we turn the key a huge amount of current will flow from solenoid coils to starter motor which will rotate the
pinion which is attached to engine later engine cranking occurred and car will start. After the car has started the plunger will comes
back and supply to the starter motor will be turned OFF. During starting huge current demanded by starting system mainly starter motor to crank the engine. This huge amount for current is
supplied by 12V lead acid battery, so the battery SOC is reduced. Not only for starting battery will also support the electrical
requirements when the engine is off so the charge of battery will be keep falling. So, the battery cannot serve continuously for the
electrical loads in car, So, there should be some way to recharge the battery and power to all the electrical loads such as various
lighting systems, mobile chargers and all electrical and electronic accessories in car. So, charging system serves this purpose for
charging the battery and give supply to the electrical components in car. The Charging system consist of Automotive Alternator
which is attached to engine via belt and also it contains full bridge rectifier and to convert supply from alternator from AC to DC.
This DC supply will charge the battery and also used for self excitation of its own field winding. This Automotive system is shown
in fig. 10 is same for both topologies. Automotive Alternator will keep on running in conjunction with engine. So, the output of alternator will vary continuously with the
speed so it is not desirable for this case a field voltage regulator is included in that circuit to set field voltage in limits there by
output in limits.
The proposed model will have the flexible thin-film solar PV module positioned on the roof top, supported by a Maximum Power
Point Tracking charge controller, provides the energy to recharge the car battery when the vehicle is stationary. Because the MPPT
is based on the Power Versus Voltage characteristics of a Solar Panel, the voltage of the Solar Panel may be regulated to give
maximum power to the load by inserting the MPPT DC-DC Converter between the Solar Panel and the Battery. The existing
alternator in the car's electrical system takes over the battery charging requirements when the vehicle is in motion. .In the
SIMULINK model we have included the fuel consumption display system.
It consists of all the possible loads in automotive, it consists of starter motor, solenoid, ignition system, lighting systems,
Fuel pump motor is modelled as equivalent resistance and also hybrid energy system using PV array with constant irradiance and
with variable irradiance values, and fuel indication and BMS. A MPPT is used for extracting the maximum power from the solar PV module and transferring that power to the load. A Buck
converter serves the purpose of transferring maximum power from the solar PV module to the load. A Buck converter acts as an
interface between the load and the module. By changing the duty cycle the load impedance as seen by the source is varied and
matched at the point of the peak power with the source so as to transfer the maximum power. Therefore, a MPPT technique is
required to obtain maximum power from a PV system. MPPT is used in photovoltaic (PV) systems to maximize the photovoltaic
array output power, irrespective of the temperature and radiation conditions and of the load electrical characteristics with the use of
DCDC converter like buck converter configuration.
Increase in electrical and electronic accessories in automotive system and increasing fuel prices and environmental concerns provide
automotive industry to produce more efficient vehicles. However, because to the constraints imposed by the energy supply of
existing Lead acid battery, there are still some questions about its performance and dependability. In this we have developed a
battery-Rooftop PV array Hybrid Energy System to supply power to AES in a best efficient way. The goal is to reduce the fuel
consumed by the engine to provide supply to the automotive loads by hybrid operation of rooftop PV array. In these various
topologies are included for storing of PV power when we have constant irradiance of 1000W/m^2.Here we have observed that it has
costed around 35 RS/hr for existing regular AES system to supply its loads which is around 350 RS/hr for 10hours of operation of
vehicle which is higher than the conventional power generation for unit cost. So with the proposed hybrid energy system for AES
system, we have got fuel consumption around 27 RS/hr and it is around 270 RS/hr for 10 hours of operation which is less than
convention system as and also we know the solar power is free of cost. By integrating the solar panels less energy from ICE is
demanded. With a smaller engine system, the overall efficiency of AES system gains a significant improvement. As a result, fuel
consumption and pollutant emission can be reduced.
"""

second_sen = """
A modern automobile is no longer simple mechanical
equipment, but a complex system that integrates hundreds of
Electronic Control Units (ECUs) connecting through the
Controller Area Network (CAN) bus [1]. The CAN bus has
become the de facto in-vehicle bus standard because it is
designed with full consideration of the real-time
communication. However, the lack of security mechanisms
such as authentication, encryption, and integrity verification
results in serious cyber-security risks in the CAN bus.
Moreover, intelligent automobiles are becoming more
dependent on the Internet, which exposes various types of
attack interfaces, including Bluetooth, Wi-Fi, and cellular
networks, for the attackers to intrude in the in-vehicle networks
[2]. Once the attackers successfully access the in-vehicle
networks through the CAN bus, they can inject arbitrary frames
to compromise the security- and safety-critical ECUs.
In order to address the security issues in automobiles,
researchers have proposed several methods for the CAN bus
security test. Penetration testing is a commonly used technique
for mining the cyber-security vulnerabilities of the CAN bus.
Based on the degree of acquisition of relevant information
about the tested object, penetration testing can be divided into
three categories, i.e., the white-box test, the gray-box test, and
the black-box test. Fuzz testing is a typical black-box test, which can launch automated security test on the target system
without knowledge about the internal operating mechanism of
the tested object. Therefore, fuzz is a promising method to
provide security protection for the CAN bus of automobiles [3].
The major challenges for the fuzz testing in the automotive
CAN bus are three-folds. Firstly, in the application layer of the
CAN bus protocol, the communication matrix is the
manufacturer's private information [4]. Without the
communication matrix, the fuzzy data generation algorithms
can only generate random fuzzy data in the CAN bus data field,
which is inefficient and time-consuming due to the data
explosion. Secondly, the generated abnormal fuzzy frames that
deviate significantly from the normal frames are very likely to
be filtered out by the vehicular Intrusion Detection System
(IDS). Thus, it is difficult for the mutation strategy to generate
effective fuzzy data. Thirdly, how to monitor the slight
anomalies of the CAN bus in the fuzz testing is also a very
challenging issue. Most of the firmware in ECU is relatively
simple, even without an operating system. In addition, the
"challenge-response" based access control mechanism makes it
difficult to install the anomaly monitoring program on an ECU
to access the logs for the anomaly analysis.
In this paper, we propose CAN-FT, a fuzz testing method
for mining and analyzing the cyber-security vulnerabilities in
automotive CAN bus. The major contributions can be
summarized as follows:
First, we adopt the reverse engineering based on the Bit
Flip Rate (BFR) and the Generative Adversarial Network
(GAN) for the fuzzy data generation algorithm, which is able to
efficiently generate effective fuzzy data without the
communication matrix.
Second, we propose the Adaptive Boosting (AdaBoost)
based approach to provide the fine-grained anomaly monitoring
capability for the fuzz testing. It guarantees that the slight
anomalies causing no obvious behaviors can be captured and
analyzed. Third, we extensively evaluate the performance of the
proposed CAN-FT in a set of indoor experiments and in a
brand-new vehicle. The experimental results show that
compared with the existing fuzz testing methods, CAN-FT is
able to find vulnerabilities more efficiently and
comprehensively.
The CAN bus is a multi-host broadcast bus protocol with
large bandwidth, anti-interference, and real-time
communication capabilities, which makes it been widely used
in in-vehicle networks. As shown in Figure 1, according to the
length of the CAN ID in the arbitration field, CAN data frames
are divided into standard frames and extended frames. The
standard data frame is composed of 7 major fields. The start of
data frame (SOF) informs each receiver of the start of data
transmission. The arbitration field prevents the data frame from
colliding during transmission. The data field transmits up to 8
bytes of data. Cyclic redundancy code is used to identify
whether the data frame is normally sent to the receiver. If the
receiver receives the data frame normally, the ACK bit will be
activated. The data frame end bit indicates that the transmission
process of the data frame has ended. The control domain
resolves the bus data competition. The data length field (DLC)
defines the length of the data field. Unfortunately, due to the proprietary application layer
protocol, we are unable to clarify what the 64-bit data field
means. According to existing research results [5-7], the 64-bit
data field is divided into multiple segments as following.
1) Signals. Values of vehicular status. As some physical
signals change with changes in roads and driving conditions,
the changes in signal segments are unpredictable, such as signal
segments that indicate speed.
2) Multi-Values. A few fixed values that are used to
represent several states of a certain signal of the vehicle, such
as the opening and closing of the door.
3) Constants. Values that do not change over time. The
segment may not be used, or they may represent a fixed value.
4) Counters. Signals that behave as cyclic counters within a
specific range. Counters could serve as additional syntax
checks [7] or be intended to prevent replay attacks.
5) Check codes. The segment is usually located at the end
of the data field and is used to verify the integrity of the data.
As information systems have become more and more
complex, traditional penetration testing has become more
energy and time consuming. Automated security test
technology that can be used to discover vulnerabilities has
received more and more attention. Fuzz is an automated
security test technology in software and information system.
The Fuzz tool repeatedly runs a program with random,
unexpected or malformed fuzzy data as input to discover
vulnerabilities.
A fuzz framework must consist of two core modules: fuzzy
data generator, and anomaly monitor. The fuzzy data generator
generates the fuzzy input for the object under test. However,
the input space of object under test is too large to be
enumerated, which causes that smarter fuzzy data generation
method is needed to avoid input explosion. Besides random
generation method, the fuzzy data can be generated based on a
mutation strategy on an existing real input. Random generation
method is quite often very ineffective with the low probability
of generating valid input. Mutation based input generation
method uses an existing real input which is modified in some
way. And the modification can be as simple as flipping some
random bits, but can also include adding or removing data.
The anomaly monitor is responsible for determining
whether the given execution of an object has a bug or anomaly
by determining whether the given execution violated a security
policy.
As shown in Figure 2, CAN-FT consists of two core
modules: data generator and anomaly monitor. The data
generator uses two algorithms to generate fuzzy data for
mining the potential vulnerabilities of the CAN bus, i.e., the
generator based on Bit Flip Rate (BFR) and the generator based
on Wasserstein GAN with Gradient Penalty (WGAN-GP). The
generated fuzzy data should conform to the frame structure of
the CAN bus protocol to avoid being ignored by ECUs. The
anomaly monitor is used to monitor the abnormal behavior of
the tested CAN bus during the fuzz testing. The fuzzy data
causing the anomalies will be captured to mine the potential
vulnerabilities of the CAN bus.
A random mutation strategy for the 8-byte data field will
cause data explosion. Moreover, the fuzzy data generated by
the random mutation strategy will deviate significantly from
the data in the normal state. If the value of a segment exceeds a
certain threshold, the fuzzy data will be intercepted by the
vehicular IDS, which is ineffective and impact the efficiency of
fuzz testing.
Thus, we propose the BFR based fuzzy data generator. At
the CAN bus application layer, CAN bus messages are
composed of ID domain and 8-byte long data domain. The
BFR is the frequency at which a certain bit in the data field
changes in different CAN messages with the same ID over a
period. Assuming that the CAN bus message is expressed as message data field with ID g at time . The greater the
, the number of changes in time is more. There are
five kinds of segments in the data field of CAN messages for
different purposes [7], which causes significant differences in
the BFR of bits in different segments. And we can use the
difference between the BFRs to segment the data field of the
CAN messages. The BFR determines the type of segment, and
helps us to roughly infer the role of the segment in the data
field.
Obviously, the BFR of the constant field is 0. The BFR of
the multi-value segment is random with no fixed change rule.
The user triggers a certain vehicular behavior, which causes the
states of the vehicle to change. As a result, the multi-value field
changes, causing the change of BFR. The value of the counter
segment is incremented. Therefore, the lowest bit of the
counter segment has the highest BFR, and each subsequent bit
will be twice as low as the previous bit. The BFR of the check
code segment that is located at the end of the data field obeys a
normal probability distribution centered at 0.5 [6]. Based on the observation, the normality test can be performed on the BFR of
check code segment. If it conforms to the normal distribution,
the segment can be considered as a check code segment. Due to
the limitation of the cycle time, the flip rate between two
consecutive bits in the signal segment will be smaller. And the
lower bit has a higher flip rate than the higher bit, which causes
that the BFR between different signal boundaries will vary
greatly.
With the knowledge of BFR in different segments, the data
field of CAN messages can be segmented according to three
rules. First, the variable and immutable region can be extracted
based on whether the BFR of each bit changes. The immutable
region is regarded as constant segment. Second, the change of
the BFRs of two adjacent bits in the signal segment is checked
in turn. When the BFR of the second bit divided by the BFR of
the first bit is less than a preset threshold, a preliminary
boundary is divided between the two bits. Third, the
preliminary segment division of the previous stage will be
analyzed again. The variable region will be classified according
to the BFRs of the check code segment, the counter segment,
and the multi-value segment. If the region is not these types of
segments, mark the region as a signal segment. Based on the
segments obtained, the data generator in CAN-FT can finely
control data mutation strategies, adopting different mutation
strategies for different segments based on their characteristics. GAN [8] is a deep learning algorithm composed of a
generator and a discriminator that generally adopt a neural
network structure. The generator selects random noise as input
for training, while the discriminator uses real data and the fuzzy
data from the generator for training. According to the feedback
of the discriminator to the data, the parameters of the generator
and the discriminator will be continuously adjusted. Through
the game between them, the fuzzy data of the generator will
gradually approach the real data.
However, GAN has problems such as model nonconvergence, mode collapse, gradient disappearance, and
gradient explosion. To solve the above problems, Wasserstein
GAN (WGAN) model is firstly proposed in [9]. Gulrajani et al.
[10] proposed WGAN-GP that improved WGAN with gradient
penalty and removed the weight constraint to accelerate the
training process.
1) Dataset. The collected message through CAN transceiver
is usually displayed in hexadecimal format. In order to create a
dataset for training and generating a GAN network, redundant
and repeated messages will be deleted. Both hexadecimal data
field and ID field will be converted to binary vector that will be
spliced together, constituting an input vector in the dataset.
2) Model Training. The generator and discriminator of
WGAN-GP are constructed using a fully connected neural
network. The generator includes an input layer, three hidden
layers, and an output layer. The Rectified Linear Unit function
is selected as the activation function of the hidden layer to
avoid gradient disappearance in the back propagation process.
The generator selects noise that obeys a normal distribution as
input. According to the Wasserstein theory, stochastic gradient
descent method is used to optimize the objective function. The discriminator uses the same neural network structure as
generator but removes the sigmoid activation function of the
last layer. CAN messages from the On-Board Diagnostics (OBD)
interface of vehicle are used to generate the training dataset.
Then the generator and the discriminator are continuously
optimized. The final generator can generate fuzzy messages
conforming to the protocol specification that are close to real
message from physical vehicle. Adaptive Boosting (AdaBoost) [11] is a boosting method
that combines multiple weak classifiers into a strong classifier.
AdaBoost first trains an initial weak classifier, calculates its
classification error rate, and determines the weight of the
classifier. In the training process, if a sample has been
accurately classified, its weight will be reduced in the
construction of the next training set. On the contrary, if a
sample is not accurately classified, its weight will be increased.
Then, the sample set after updating weights is used to train the
next classifier. The whole training process proceeds iteratively
in this way until the classification error rate decreases to a
certain value or the number of iterations reaches a specified
value. Finally, according to the weights of all classifiers, it can
be combined into a strong classifier.
The paper implements an anomaly monitor based on
AdaBoost algorithm. The collected real messages from the
CAN bus and fuzzy abnormal messages are used to train
AdaBoost based CAN bus abnormal state classifier, which
detects the states of the CAN bus to determine whether it is an
anomaly in CAN bus. The procedure of proposed fuzz method is depicted in
Figure 2. The entire test process is divided into four phases:
data acquisition, training and generation, data injection, and
state monitoring.
1) Data Acquisition. The data acquisition consists of step 1
and step 2 in Figure 2. Before starting fuzz testing, CAN-FT
uses the CAN bus transceiver connected between the test
machine and the automobile to collect CAN messages in the
CAN bus, which will be used to train the GAN and the
anomaly monitor.
2) Classifier Training and Fuzzy Data Generation. Step 3-1
in Figure 2 shows the CAN messages processing procedure in
the data generator based on BFR. Step 3-2 describes the
procedure of data generator based on WGAN-GP. And in step
3-3, AdaBoost based classifier in CAN-FT is trained with both
normal CAN messages from automobile and fuzzy abnormal
CAN messages.
3) Data Injection. As illustrated in step 4 and step 5, CANFT needs to inject the generated fuzzy data into the target CAN
bus with the help of the CAN bus transceiver. The injected
messages may cause the abnormal behaviors of the automobile
if there are vulnerabilities in the CAN bus.
4) State Monitoring. After model training in step 3-3, the
anomaly state classifier is ready. The anomaly monitor
identifies the CAN bus state based on AdaBoost after the fuzzy
data are injected into the automobile. 
"""
```
