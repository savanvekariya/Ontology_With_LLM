```python
import IPython
from pyvis.network import Network


def save_network_html(kb, nodes):
    # create network
    net = Network(
        directed=True, width="1500px", height="900px", bgcolor="#eeeeee", notebook=True
    )

    # nodes
    color_entity = "#00FF00"
    for e in nodes:
        net.add_node(e, shape="circle", color=color_entity)

    # edges
    for r in kb:
        net.add_edge(r["head"], r["tail"], title=r["relation"], label=r["relation"])

    # save network
    net.repulsion(
        node_distance=200,
        central_gravity=0.2,
        spring_length=200,
        spring_strength=0.05,
        damping=0.09,
    )
    net.set_edge_smooth("dynamic")
    net.show("doc3.html")
```

```python
save_network_html(triplets, nodes)
```
